### 3 There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?

```python
# I guess it is the policy ? We could group them in order to get only the general types instead of every types.


import numpy as np

customer_df["Policy"] = np.where(customer_df["Policy"].str.contains("Personal"), "Personal", 
                    np.where(customer_df["Policy"].str.contains("Corporate"), "Corporate", 
                    np.where(customer_df["Policy"].str.contains("Special"), "Special", 
                             customer_df["Policy"])))

customer_df.value_counts("Policy")
```
